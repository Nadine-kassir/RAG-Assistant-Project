```python
#Hugging Face embedding model
huggingface_embeddings = HuggingFaceBgeEmbeddings(
    model_name= "sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs = {'device' : 'cuda'},
    encode_kwargs = {'normalize_embeddings' : True}
)
```

```python
#turning FAISS vector store into a retriever object, which will be plugged directly into RAG pipeline.
retriever = vectorstore.as_retriever(search_type="similarity" , search_kwargs={"k" : 3})
```

```python
query = "What are the main risk factors for breast cancer?"
docs = retriever.get_relevant_documents(query)

for i, doc in enumerate(docs):
    print(f"\nDocument {i+1}:\n")
    print(doc.page_content[:500])  # show first 500 chars
```

```python

# connect a model from Hugging Face to generate text answers from the retrieved context.
model_id = "mistralai/Mistral-7B-Instruct-v0.2"

pipe = pipeline(
    "text-generation",
    model=model_id,
    max_new_tokens=512,
    temperature=0.3,
    repetition_penalty=1.1
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
prompt_template = """
You are an expert medical assistant.
Use the following pieces of context to answer the question accurately and clearly.
If you don't find enough information, say "I don't have enough data to answer confidently."

Context:
{context}

Question: {question}

IMPORTANT: You must wrap your entire answer inside <results></results> tags.

Format your response exactly as follows:
<results>
[Your answer here]
</results>

Answer:
"""

PROMPT = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
```

```python
# ✅ Create the RetrievalQA chain correctly
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # tells LangChain how to use your prompt
    retriever=vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3}),
    return_source_documents=True ,
    chain_type_kwargs={"prompt": PROMPT}

)
```

```python
import re

def classify_intent(user_input):
    """Classify user intent before running RAG"""
    user_input = user_input.lower().strip()

    # Greetings
    greetings = ['hello', 'hi', 'hey', 'good morning', 'good afternoon', 'good evening']
    if any(greet in user_input for greet in greetings):
        return "greeting"

    # Farewells
    farewells = ['bye', 'goodbye', 'see you', 'thanks', 'thank you']
    if any(word in user_input for word in farewells):
        return "farewell"

    # Breast cancer related keywords
    bc_keywords = ['breast', 'cancer', 'mammogram', 'tumor', 'treatment',
                   'chemotherapy', 'radiation', 'biopsy', 'mastectomy', 'lumpectomy']
    if any(keyword in user_input for keyword in bc_keywords) or '?' in user_input:
        return "domain_query"

    # If very short and doesn't seem like a question
    if len(user_input.split()) <= 3 and '?' not in user_input:
        return "casual"

    return "unknown"

def extract_results(response):
    """Extract content from <results> tags - finds LAST occurrence"""
    start_tag = '<results>'
    end_tag = '</results>'

    start_indices = [i for i in range(len(response)) if response.startswith(start_tag, i)]
    end_indices = [i for i in range(len(response)) if response.startswith(end_tag, i)]

    if start_indices and end_indices:
        start_idx = start_indices[-1] + len(start_tag)
        end_idx = end_indices[-1]

        if end_idx > start_idx:
            content = response[start_idx:end_idx].strip()
            print(f"DEBUG: Extracted from last occurrence, length: {len(content)}")
            return content

    return response.strip()

def query_rag_fixed(query):
    """Fixed version with intent classification"""
    try:
        # ✅ ADD INTENT CLASSIFICATION HERE - BEFORE RAG
        intent = classify_intent(query)

        # Handle non-domain queries without using RAG
        if intent == "greeting":
            answer = "Hello! 👋 I'm your breast cancer information assistant. I can help answer questions about breast cancer, treatments, symptoms, prevention, and more. What would you like to know?"
            print("\n" + "=" * 80)
            print("ANSWER:")
            print(answer)
            print("=" * 80)
            return answer

        elif intent == "farewell":
            answer = "Goodbye! Take care and feel free to return if you have more questions about breast cancer. 💙"
            print("\n" + "=" * 80)
            print("ANSWER:")
            print(answer)
            print("=" * 80)
            return answer

        elif intent == "casual":
            answer = "I'm here and ready to help! I specialize in breast cancer information. Would you like to ask me something about breast cancer?"
            print("\n" + "=" * 80)
            print("ANSWER:")
            print(answer)
            print("=" * 80)
            return answer

        elif intent == "unknown":
            answer = "I'm specialized in providing information about breast cancer. Could you please ask me a question related to breast cancer topics like symptoms, diagnosis, treatment, or prevention?"
            print("\n" + "=" * 80)
            print("ANSWER:")
            print(answer)
            print("=" * 80)
            return answer

        # Only proceed with RAG for domain queries
        print(f"🔍 Running RAG for domain query...")
        result = qa_chain.invoke({"query": query})
        full_result = result.get('result', 'No answer found.')

        # Extract answer (will use LAST occurrence)
        answer = extract_results(full_result)

        # Display
        print("\n" + "=" * 80)
        print("ANSWER:")
        print(answer)
        print("=" * 80)

        # Sources
        print("\n📚 SOURCES:")
        source_docs = result.get('source_documents', [])
        sources = {}
        for doc in source_docs:
            src = doc.metadata.get('source', 'Unknown')
            page = doc.metadata.get('page', 'N/A')
            sources.setdefault(src, []).append(page)

        for i, (src, pages) in enumerate(sources.items(), 1):
            page_list = ', '.join(map(str, sorted(set(pages))))
            print(f"  {i}. {src} (Pages: {page_list})")

        print("=" * 80)

        return answer

    except Exception as e:
        print(f"❌ Error: {str(e)}")
        return None

# Test with different types of inputs
print("Test 1: Greeting")
answer = query_rag_fixed("hello")

print("\n\nTest 2: Domain query")
answer = query_rag_fixed("What are the latest treatments for breast cancer?")

print("\n\nTest 3: Casual")
answer = query_rag_fixed("hi there")
```

```python
import gradio as gr
import re

# NEW: Add intent classification function
def classify_intent(user_input):
    """Classify user intent before running RAG"""
    user_input = user_input.lower().strip()

    # Greetings
    greetings = ['hello', 'hi', 'hey', 'good morning', 'good afternoon', 'good evening']
    if any(greet in user_input for greet in greetings):
        return "greeting"

    # Farewells
    farewells = ['bye', 'goodbye', 'see you', 'thanks', 'thank you']
    if any(word in user_input for word in farewells):
        return "farewell"

    # Breast cancer related keywords
    bc_keywords = ['breast', 'cancer', 'mammogram', 'tumor', 'treatment',
                   'chemotherapy', 'radiation', 'biopsy', 'mastectomy', 'lumpectomy',
                   'symptom', 'diagnosis', 'screening', 'prevention', 'risk']
    if any(keyword in user_input for keyword in bc_keywords) or '?' in user_input:
        return "domain_query"

    # If very short and doesn't seem like a question
    if len(user_input.split()) <= 3 and '?' not in user_input:
        return "casual"

    return "unknown"

# Predefined extraction function - finds LAST occurrence of <results> tags
def extract_results(response):
    """Extract content from LAST occurrence of <results> tags"""
    start_tag = '<results>'
    end_tag = '</results>'

    # Find LAST occurrence using rfind()
    end_idx = response.rfind(end_tag)

    if end_idx != -1:
        # Find the <results> that comes before this </results>
        start_idx = response.rfind(start_tag, 0, end_idx)

        if start_idx != -1:
            content = response[start_idx + len(start_tag):end_idx].strip()
            if content:  # Make sure it's not empty
                return content

    # Fallback: return full response if no tags found
    return response.strip()

# MODIFIED: Wrap RAG chatbot in a function with intent classification
def ask_bot(question):
    if not question or question.strip() == "":
        return "Please enter a question."

    try:
        # ✅ ADD INTENT CLASSIFICATION HERE
        intent = classify_intent(question)

        # Handle non-domain queries without using RAG
        if intent == "greeting":
            return "Hello! 👋 I'm your breast cancer information assistant. I can help answer questions about breast cancer, treatments, symptoms, prevention, and more. What would you like to know?"

        elif intent == "farewell":
            return "Goodbye! Take care and feel free to return if you have more questions about breast cancer. 💙"

        elif intent == "casual":
            return "I'm here and ready to help! I specialize in breast cancer information. Would you like to ask me something about breast cancer?"

        elif intent == "unknown":
            return "I'm specialized in providing information about breast cancer. Could you please ask me a question related to breast cancer topics like symptoms, diagnosis, treatment, or prevention?"

        # Only proceed with RAG for domain queries
        # Use invoke method for the RAG chain
        result = qa_chain.invoke({"query": question})

        # Get the full result
        full_result = result.get('result', 'No answer found.')

        # Extract clean answer using predefined function
        clean_answer = extract_results(full_result)

        # Get sources
        source_docs = result.get('source_documents', [])
        if source_docs:
            sources_text = "\n\n📚 **Sources:**\n"
            sources = {}
            for doc in source_docs:
                src = doc.metadata.get('source', 'Unknown')
                page = doc.metadata.get('page', 'N/A')
                sources.setdefault(src, []).append(page)

            for i, (src, pages) in enumerate(sources.items(), 1):
                page_list = ', '.join(map(str, sorted(set(pages))))
                sources_text += f"{i}. {src} (Pages: {page_list})\n"

            return clean_answer + sources_text
        else:
            return clean_answer

    except Exception as e:
        return f"❌ Error: {str(e)}\n\nPlease try rephrasing your question."

# Create Gradio interface with enhanced features
interface = gr.Interface(
    fn=ask_bot,
    inputs=gr.Textbox(
        label="Your Question",
        placeholder="e.g., What are the risk factors for breast cancer?",
        lines=3
    ),
    outputs=gr.Textbox(
        label="Answer",
        lines=12
    ),
    title="🩺 Breast Cancer RAG Chatbot",
    description="Ask questions about breast cancer based on the loaded medical PDFs. The system uses RAG to provide accurate, context-based answers.",
    examples=[
        ["What are the main types of breast cancer?"],
        ["What are the early symptoms of breast cancer?"],
        ["What screening methods are recommended?"],
        ["What are the treatment options available?"],
        ["Hello"],  # Added example to test greeting
    ],
    theme=gr.themes.Soft(),
    allow_flagging="never"
)


# Launch - simplified for Colab
interface.launch(share=True)
```
